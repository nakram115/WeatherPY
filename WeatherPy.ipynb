```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
style.use('fivethirtyeight')
import requests
import json
import citipy as citipy
import openweathermapy.core as owm
import pickle 
import csv
```

```python
# Import cities file as DataFrame
cities_pd = pd.read_csv("worldcities.csv")
cities_pd.head()
```

```python
random_cities=cities_pd.sample(550,replace=True)
random_out=open("random.pkl",'wb')
pickle.dump(random_cities,random_out)
random_out.close()
```

```python
random_in=open("random.pkl","rb")
random_cities=pickle.load(random_in)
random_cities.head()
```

```python
random_cities["Temperature (F)"] = ""
random_cities["Humidity"] = ""
random_cities["Cloudiness"] = ""
random_cities["Wind Speed (mph)"] = ""
random_cities.head()
```

```python
my_tups=[]
log_file=open('logfile.log','w')

for row in random_cities.itertuples():
    country_,city_,lat_,long_=getattr(row,"Country"),getattr(row,"City"),getattr(row,'Latitude'),getattr(row,"Longitude")
    
    url="http://api.openweathermap.org/data/2.5/weather?lat={}&lon={}&appid={}".format(lat_,long_,"adabf952ce676f5038cb9a0e2556e9c9") 
    info=  requests.get(url).json()
    log_file.write("City:{}, url:{}\r\n".format(city_,url))
    
    temp=(info["main"]["temp"]-273)*(9/5)+32
    humidity=info["main"]["humidity"]
    clouds=info["clouds"]["all"]
    wind=info["wind"]["speed"]
    
    my_tups.append((country_,city_,lat_,long_,temp,humidity,clouds,wind))
 
log_file.close()
```

```python
new_df=pd.DataFrame(my_tups,columns=['Country','City','Latitude','Longitude','Temperature','Humidity','Cloudiness','Wind Speed'])
new_df.head()
new_df.to_csv("new_df.csv")
```

```python
# Make scatterplots
plt.scatter(x=new_df['Latitude'], y=new_df['Temperature'], marker="o", facecolors="red", edgecolors="black", alpha=0.75)
plt.title("Temperature(F) VS Latitude")
plt.xlabel("Latitude")
plt.ylabel("Temperature (F)")
fig=plt.gcf()
plt.show()
fig.savefig("Lat_Temp.png")
```

```python
plt.scatter(x=new_df['Latitude'], y=new_df['Humidity'], marker="o", facecolors="blue", edgecolors="black", alpha=0.75)
plt.title("Humidity (%) VS Latitude")
plt.xlabel("Latitude")
plt.ylabel("Humidity (%)")
fig=plt.gcf()
plt.show()
fig.savefig("Lat_Hum.png")
```

```python
plt.scatter(x=new_df['Latitude'], y=new_df['Cloudiness'], marker="o", facecolors="red", edgecolors="black", alpha=0.75)
plt.title("Cloudiness(%) VS Latitude")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness (%)")
fig=plt.gcf()
plt.show()
fig.savefig("Lat_Cloud.png")
```

```python
plt.scatter(x=new_df['Latitude'], y=new_df['Wind Speed'], marker="o", facecolors="blue", edgecolors="black", alpha=0.75)
plt.title("Windspeed(MPH) VS Latitude")
plt.xlabel("Latitude")
plt.ylabel("Windspeed (MPH)")
fig=plt.gcf()
plt.show()
fig.savefig("Lat_Wind.png")
plt.show()
```
